```python
import torch
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_id = "meta-llama/Llama-3.2-1B-Instruct"
```

```python
pipe = pipeline(
    "text-generation", 
    model=model_id, 
    torch_dtype=torch.bfloat16, 
    device_map="auto"
)
pipe("The key to life is")
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",
    load_in_4bit=True 
)
```

```python
def build_prompt(text):
    return f"""
You are a professional English to Hindi translator.

Rules:
- Translate English to Hindi
- Output ONLY Hindi translation
- Use ONLY Devanagari script
- Do NOT use Roman Hindi
- Do NOT explain anything

English: {text}
Hindi:
"""
```

```python
def translate_en_hi(text):
    prompt = build_prompt(text)

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=80,
            do_sample=False,
            temperature=0.0,
            repetition_penalty=1.25,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.eos_token_id
        )

    # ✅ decode only new tokens
    new_tokens = outputs[0][inputs["input_ids"].shape[1]:]
    result = tokenizer.decode(new_tokens, skip_special_tokens=True).strip()

    # ✅ post-clean
    result = result.split("\n")[0]
    return result
```

```python
print(translate_en_hi("Machine translation helps low resource languages."))
```

```python
print(translate_en_hi("The weather is very pleasant today."))
```
